### Principles of good/effective visualization

<b>An effective visualization is one that conveys the correct understanding to the audience with clarity and simplicity, enabling them to take useful actions.</b> 

According to David McCandles, a British data journalist (http://davidmccandless.com/), there are four elements in an effective data visualization:

- The Information: The pieces of data that you wish to share
- The Story: story is a clear and compelling narrative that ties the data and your experiences around it together
- The Goal: Goal is the objective for creating the visual
- The Visual Form: A visual expression that represents the information

#### Design Principles for effective visualization

1. Choose the right type of chart/graph that conveys your message as clearly as possible. Prefer the simple graphs over the complicated ones. 

2. Optimize the data-ink ratio. Focus on the areas of the graphs that convey essential message of the chart. Minimize non-essential use of ink in chart e.g., shadows, boxes aroung legends etc.

3. Use orientation effectively to enhance readability of your chart.

4. Use colors cautiously, use consistent and inclusive color schemes.

5. Minimize the clutter - keep the number of elements like lines, segments in your chart to just what is required to convey the essential message. 

6. Avoid false, misleading and deceptive visuals. 

#### Challenges with data visualization.

- Since visualizations have a higher influence on us, they also have a higher potential to mislead us through bias and misinformation. 
- For example, a slant in the title or an inconsistency between title and the details of the visualization may indicate that the information is biased. Text elements within a graphic may also convey misinformation. 

We can look for some of these features in a graphic to identify if it is manipulative [3]: 

1. Y-axis does not start at 0. By changing the y-axis scale, the differences between the graphs like bar charts can be made to appear bigger or smaller than they are.
2. Dual Y-axis can be misleading.
3. Choosing only the desired parts of data. 
4. Misleading grouping of data.
5. Non-monochromatic color scale for a quantitative variable.
7. Sum of percentages in slices of pie-chart is not equal to hundred.
8. Use of a cumulative chart can disguise some insightful trends
9. Inconsistency in the title of the visualization and its content. [12]
10. Use of smooth trend lines in a scatter plot can mislead the audience about the real trends.

The Story that you tell with a visualization may easily be just the fictious story you would like to tell. <b>The story you tell must be a story that conveys facts. </b> 

## Visualizing Climate Change in Asia since the year 1950

Data Sources: This dataset is a part of data compiled by BerkeleyEarth (BerkeleyEarth.org). It contains average temperature in degree celcius for different countries of the world over a period of years from 1743 to 2013. 
We have downloaded this CSV file from Kaggle. 

Dataset URL: https://www.kaggle.com/datasets/berkeleyearth/climate-change-earth-surface-temperature-data?resource=download&select=GlobalLandTemperaturesByCountry.csv    

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp = pd.read_csv("GlobalLandTemperaturesByCountry.csv", parse_dates = ["dt"])
```

```python
temp.head()
```

```python
temp.shape
```

```python
temp.dtypes
```

```python
temp = temp.dropna()
temp.shape
```

```python
temp.head()
```

```python
# Remove the country 'Aland'
temp = temp.query("Country != 'Åland'")
temp.shape
```

```python
#add a new column 'Year' to dataframe, we extract this from original date column
temp['Year'] = pd.DatetimeIndex(temp['dt']).year
```

```python
# Drop original date colummn
temp = temp.drop(["dt"], axis = 1)
temp.head()
```

```python
# we will retain the data rows only for years 1950 onwards till the year 2013
temp  = temp.loc[temp["Year"] >= 1950]
```

```python
temp = temp.dropna()
temp.shape
```

```python
# We will only limit to Asian Region
countries = ['China', 'India', 'Indonesia', 'Iran', 'Israel', 'Japan', 'Malaysia',  
             'Qatar', 'South Korea', 'Sri Lanka', 'Thailand', 'United Arab Emirates']

temp  = temp.loc[temp["Country"].isin(countries)]
temp.Country.unique()
```

```python
temp.shape
```

```python
temp.head()
```

```python
gp_country = temp.groupby("Country")
gp_country.groups.keys()
```

```python
gp_country.get_group("China")
```

```python
mean_temp_countrywise = pd.DataFrame(gp_country['AverageTemperature', 'AverageTemperatureUncertainty'].mean()).reset_index()
mean_temp_countrywise = mean_temp_countrywise.sort_values('AverageTemperature')
mean_temp_countrywise
```

```python
import matplotlib.pyplot as plt

temperatures = mean_temp_countrywise.AverageTemperature
country = mean_temp_countrywise.Country
q1 = temperatures.quantile(0.25)
q2 = temperatures.quantile(0.5)
q3 = temperatures.quantile(0.75)


bar_colors  = [{t <q1: 'blue',q1<=t<=q2: 'green', q2 < t <= q3: 'orange',t > q3: 'red'}[True] 
               for t in mean_temp_countrywise['AverageTemperature']]

plt.rcParams["figure.figsize"] = (20, 9)
plt.bar(x = mean_temp_countrywise['Country'],  height = mean_temp_countrywise['AverageTemperature'] , color = bar_colors)
plt.ylabel("Average Temperature(Celcius)")
plt.xlabel("Country")
plt.title('Courtry-wise Average Temperatures For the Years 1950 to 2013' , )
plt.show()
```

```python
gp_year = temp.groupby("Year")
gp_year.groups.keys()
```

```python
mean_temp_yearwise = pd.DataFrame(gp_year['AverageTemperature', 'AverageTemperatureUncertainty'].mean()).reset_index()
mean_temp_yearwise = mean_temp_yearwise.sort_values('AverageTemperature')
mean_temp_yearwise
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

temperatures = mean_temp_yearwise.AverageTemperature
year = mean_temp_yearwise.Year
q1 = temperatures.quantile(0.25)
q2 = temperatures.quantile(0.5)
q3 = temperatures.quantile(0.75)
min_t = temperatures.min()
max_t = temperatures.max()
min_y = year.min()
max_y = year.max()

#print(q1, q2, q3, min_t, max_t)
bar_colors  = [{t <q1: 'blue',q1<=t<=q2: 'green', q2 < t <= q3: 'orange',t > q3: 'red'}[True] 
               for t in mean_temp_yearwise['AverageTemperature']]

plt.rcParams["figure.figsize"] = (20, 9)
plt.bar(x = mean_temp_yearwise['Year'],  height = mean_temp_yearwise['AverageTemperature'] , color = bar_colors)
plt.xlim((min_y - 2, max_y+2))
plt.xticks(np.arange(min_y, max_y+3, 3.0))
plt.xlabel("Year")

plt.yticks(np.arange(0, max_t, 2.0))
plt.ylabel("Average Temperature(Celcius)")

plt.title('Year-wise Average Temperatures For Asian Countries for the Years 1950 to 2013' , )
plt.show()
```

## Exploratory Data Analysis for House Price Prediction for Boston Housing Dataset
DataSet: https://archive.ics.uci.edu/ml/machine-learning-databases/housing/

Data Description
 - 1. Title: Boston Housing Data, it contains fields about housing values in suburbs of Boston
 - 2. Number of Instances: 506
 - 3. Attributes: Total 14, 13 continuous attributes (including "class" attribute "MEDV"), 1 binary-valued attribute.
    - CRIM:      per capita crime rate by town
    - ZN:        proportion of residential land zoned for lots over 25,000 sq.ft.
    - INDUS:     proportion of non-retail business acres per town
    - CHAS:      Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
    - NOX:      nitric oxides concentration (parts per 10 million)
    - RM:        average number of rooms per dwelling
    - AGE:       proportion of owner-occupied units built prior to 1940
    - DIS:       weighted distances to five Boston employment centres
    - RAD:       index of accessibility to radial highways
    - TAX:      full-value property-tax rate per \$10,000
    - PTRATIO:  pupil-teacher ratio by town
    - B:        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
    - LSTAT:    % lower status of the population
    - MEDV:     Median value of owner-occupied homes in $1000's

```python
# Identify features highly correlated with median value of the house
correlations = df.corr()['medv'][:-1]
corr_df = pd.DataFrame(correlations)
corr_df
```

```python
def correlation_condition(cor):
    if -0.5 < cor < 0.5:
        return "Low"
    elif 0.5 < cor < 0.6 or -0.6 < cor < - 0.5:
        return "Medium"
    else:
        return "High"
    

def apply_style(styler):
    styler.format(correlation_condition)
    styler.background_gradient(axis=0) 
    return styler

corr_df.style.pipe(apply_style)
```

```python
highcorrelation = correlations[abs(correlations>0.5)].sort_values(ascending = False)
highcorrelation
```

```python
plt.figure(figsize=(12, 10))
corr = df.drop('medv', axis=1).corr() 
sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='Reds', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True)
```

```python
# Pairwise plots between medv and all other columns

plt.figure(figsize=(25, 25))
for i in range(0, len(df.columns), 3):
    sns.pairplot(data=df, x_vars=df.columns[i:i+3], y_vars=['medv'])
```

```python
import numpy as np
def remove_outliers(colm):
    q1=colm.quantile(0.25)
    q3=colm.quantile(0.75)
    IQR=q3-q1
    upper = colm[~(colm>(q3+1.5*IQR))].max()
    lower = colm[~(colm<(q1-1.5*IQR))].min()
    colm = np.where(colm > upper,colm.mean(), np.where(colm < lower,colm.mean(),colm))
    return colm
```

```python
for col in df.columns:
    df[col] = remove_outliers(df[col])
df.describe()
```
